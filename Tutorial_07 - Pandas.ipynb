```python
a = [1,2,3,4,5]
b = pd.Series(data=[1,2,3,4,5],index=[0,1,2,3,4])
c = {1,2,3,4,5}

print("a:",a)
print("b:",b)
print("c:",c)
```

```python
apples = pd.Series(data=[3, 2, 0, 1],index=[0,1,2,3],name="apples")
oranges = pd.Series(data=[0, 3, 7, 2],index=[0,1,2,3],name="oranges")

print(apples)
print(oranges)
```

```python
dataframe = pd.DataFrame(data={"apples":apples,"oranges":oranges}, index = apples.index)
```

```python
dataframe
```

## Drop -- (Silme İşlemi)

```python
# .drop ifadesiyle silmek istenen satırın index'i paremetre olarak verilebilir.
apples.drop(2)
```

```python
# sütun silmek için axis=1 parametresiyle sütun ismi verilerek yapılabilir.

dataframe.drop("apples", axis=1)
```

```python
# dataframe tablosundan satır silmek için axis = 0 parametresiyle satırın index numarası kullanılarak silme yapılabilir.

dataframe.drop(2, axis=0)
```

## Sort -- (Sıralama İşlemi)

```python
tablo = pd.DataFrame(data={"Tarih":["Ocak", "Mart", "Şubat", "Nisan"],
                               "Üretim Adet":[12000, 11000 , 13000 , 10000]},
                         index = [1,3,2,4])
```

```python
tablo
```

```python
# index sıralaması
tablo.sort_index()
```

```python
# sıralama 0-9
tablo.sort_values(by="Üretim Adet")
```

```python
# sıralama A-Z
tablo.sort_values(by="Tarih")
```

#### Rank ifadesi ilgili sütuna göre 1 den başlayarak sıralamaya göre rank değeri verir.

```python
tablo.rank()
```

```python
def color_negative_red(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = 'red' if val < 0 else 'black'
    return 'color: %s;' % color
```

```python
df.style.applymap(color_negative_red)
```
